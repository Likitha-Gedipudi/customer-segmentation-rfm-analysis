# 01. Data Generation

Generate synthetic e-commerce transaction data for RFM analysis.

**Output:** `data/raw/ecommerce_transactions.csv`

```python
import sys
sys.path.append('..')

import pandas as pd
import numpy as np
from src.data_generator import generate_full_dataset
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette('husl')
```

## Generate Dataset

```python
# Generate 5,000 customers with varied behavior profiles
customers_df, transactions_df = generate_full_dataset(
    n_customers=5000,
    save_path='../data/raw/ecommerce_transactions.csv'
)
```

## Quick Data Overview

```python
print("=" * 50)
print("TRANSACTION DATA OVERVIEW")
print("=" * 50)
print(f"\nTotal Transactions: {len(transactions_df):,}")
print(f"Unique Customers: {transactions_df['customer_id'].nunique():,}")
print(f"Date Range: {transactions_df['transaction_date'].min()} to {transactions_df['transaction_date'].max()}")
print(f"Total Revenue: ${transactions_df['total_amount'].sum():,.2f}")
print(f"\nColumns: {list(transactions_df.columns)}")
```

## Customer Behavior Profiles Distribution

```python
profile_dist = customers_df['behavior_profile'].value_counts()

fig, ax = plt.subplots(figsize=(10, 6))
colors = sns.color_palette('husl', len(profile_dist))
bars = ax.bar(profile_dist.index, profile_dist.values, color=colors)
ax.set_xlabel('Customer Behavior Profile', fontsize=12)
ax.set_ylabel('Number of Customers', fontsize=12)
ax.set_title('Distribution of Customer Behavior Profiles', fontsize=14, fontweight='bold')

for bar, val in zip(bars, profile_dist.values):
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 20, 
            f'{val:,}', ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.savefig('../reports/figures/customer_profiles_dist.png', dpi=150)
plt.show()
```

## Transaction Amount Distribution

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Transaction amounts
axes[0].hist(transactions_df['total_amount'], bins=50, color='steelblue', edgecolor='white', alpha=0.8)
axes[0].set_xlabel('Transaction Amount ($)', fontsize=11)
axes[0].set_ylabel('Frequency', fontsize=11)
axes[0].set_title('Distribution of Transaction Amounts', fontsize=12, fontweight='bold')
axes[0].axvline(transactions_df['total_amount'].median(), color='red', linestyle='--', label=f"Median: ${transactions_df['total_amount'].median():.2f}")
axes[0].legend()

# Category distribution
cat_counts = transactions_df['product_category'].value_counts()
axes[1].pie(cat_counts.values, labels=cat_counts.index, autopct='%1.1f%%', colors=sns.color_palette('husl', len(cat_counts)))
axes[1].set_title('Transactions by Product Category', fontsize=12, fontweight='bold')

plt.tight_layout()
plt.savefig('../reports/figures/transaction_overview.png', dpi=150)
plt.show()
```
